```python
data['Date'] = pd.to_datetime(data['Date'],errors='coerce')
```

```python
data.set_index('Date', inplace=True)
```

```python
# Calculating SImple Returns
simple_returns = data.pct_change()
simple_returns = simple_returns.dropna()
simple_returns.head(5)
```

```python
# Calculating Log Returns
log_returns  = np.log(data/data.shift(1))
log_returns = log_returns.dropna()
log_returns.head(5)
```

```python
# Calculating Portfolio Returns (Rp = w1r1 + w2r2 + w3r3) => [w1 w2 w3][r1 r2 r3]
weights = np.array([1/10, 1/10, 2/10,2/10,4/10])
portfolio_simple_returns = simple_returns.dot(weights)
portfolio_simple_returns.head(5)
```

```python
weights = np.array([1/10, 1/10, 2/10,2/10,4/10])
portfolio_log_returns = log_returns.dot(weights)
portfolio_log_returns.head(5)
```

```python
## Calculating Annualized Return using Simple Returns => (1+Avg of Simple Return)**252 -1
annualized_simple_returns = (1 + portfolio_simple_returns.mean())**252 -1
annualized_simple_returns
```

```python
## Calculating Annualized Log Return
annualized_log_returns = (portfolio_log_returns.mean() * 252)
annualized_log_returns
```

```python
## calculating Volatility
daily_volatility = np.std(portfolio_simple_returns)
annual_volatility = daily_volatility * np.sqrt(252)
annual_volatility
```

```python
# #Calculating Alpha & Beta Downloading S&P  500
# start_date = '2024-01-01'
# end_date = '2025-01-01'
# benchmark = yf.download('^GSPC', start=start_date,end=end_date)['Close']
```

```python
benchmark = pd.read_excel('Indian_Benchmark_Data_18042026.xlsx')
```

```python
benchmark.set_index('Date', inplace=True)
```

```python
benchmark.head(5)
```

```python
benchmark = benchmark.pct_change()
```

```python
benchmark = benchmark.dropna()
```

```python
benchmark.head(5)
```

```python
# Calculate Beta => beta = Covariance(Rp,Rm)/Variance(Rm)

# Converting Data Frame to a Numpy Array
combined = pd.concat(
    [portfolio_simple_returns, benchmark],
    axis=1
).dropna()

# Rename for clarity
combined.columns = ['portfolio', 'benchmark']

portfolio_returns = combined['portfolio']
benchmark_returns = combined['benchmark']

cov_matrix = np.cov(portfolio_returns, benchmark_returns)
cov_matrix
```

```python
beta = cov_matrix[0][1]/cov_matrix[1][1]
beta
# """
# beta of 0.89 indicates that suppose a market goes up by 10% then our portfolio will go up by 10 * 0.89 i.e 8.9%
# also indicates as beta < 1 it is less volatile as compared to market
# """
```

```python
# Calculating Alpha
risk_free_rate = 0.07
alpha = (np.mean(portfolio_simple_returns) - risk_free_rate/252) - beta * (np.mean(benchmark_returns) - risk_free_rate/252)
alpha = alpha * 252
alpha
```

```python
# calculating sortino ratio
negative_returns = portfolio_simple_returns[portfolio_simple_returns < 0]
downside_deviation = np.std(negative_returns)
downside_deviation = downside_deviation * np.sqrt(252)
downside_deviation
```

```python
sharpe_ratio = (annualized_simple_returns - risk_free_rate)/annual_volatility
sharpe_ratio
```

```python
"""
Sharpe ratio of 1.64 indicates that for every unit of risk taken psortfolio is generating 1.64 units of excess ratio
"""
```

```python
sortino_ratio = (annualized_simple_returns - risk_free_rate)/downside_deviation
sortino_ratio
```

```python
# Max Drawdown  = Cummulattive Return

cummulative_simple_returns = (1 + portfolio_simple_returns).cumprod()
max_drawdown = ((cummulative_simple_returns.cummax() - cummulative_simple_returns) / cummulative_simple_returns.cummax()).max()
max_drawdown
```

```python
calmar_ratio = annualized_simple_returns/max_drawdown
calmar_ratio
```

```python
treynor_ratio = (annualized_simple_returns - risk_free_rate)/beta
treynor_ratio
```
